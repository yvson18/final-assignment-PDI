```python
import cv2
import os
import numpy as np
import pandas as pd
```

```python
directory = "imgs\imgs"
h = []
w = []

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    img = cv2.imread(f)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h.append(img.shape[0])
    w.append(img.shape[1])

h = np.array(h)
w = np.array(w)
print(f"Altura da imagem --> Média: {np.mean(h)}, Mediana: {np.median(h)} e Moda:")
print(f"Largura da imagem --> Média: {np.mean(w)}, Mediana: {np.median(w)} e Moda:")
```

```python
gab = pd.read_csv("ground_truth.csv")
display(gab)
```

```python
directory = "imgs\imgs"
data = {"filename": [], "hat": [], "veil": [], "glasses": [], "accessory": []}

for i in range(len(gab)):
    f = os.path.join(directory, gab["img_name"].iloc[i])
    if os.path.isfile(f):
        #Montando o gabarito:
        data["filename"].append(gab["img_name"].iloc[i])
        data["hat"].append(gab["hat"].iloc[i])
        data["veil"].append(gab["veil"].iloc[i])
        data["glasses"].append(gab["dark_glasses"].iloc[i])
        if gab["hat"].iloc[i] == 0 or gab["veil"].iloc[i] == 0 or gab["dark_glasses"].iloc[i] == 0:
            data["accessory"].append(1)
        else:
            data["accessory"].append(0)
        #resize da imagem:
        img = cv2.imread(f)
        resized = cv2.resize(img, (1360, 1360))
        #Salvando a imagem no dataset:
        fd = os.path.join("Dataset", gab["img_name"].iloc[i])
        cv2.imwrite(fd, resized)
    else:
        pass

df = pd.DataFrame(data)
display(df)
```

```python
gab = pd.read_csv("gabarito.csv")
display(gab)
```

```python
directory = "Dataset"
data = {"filename": [], "hat": [], "veil": [], "glasses": [], "accessory": []}

for i in range(len(gab)):
    f = os.path.join(directory, gab["filename"].iloc[i])
    if os.path.isfile(f):
        if gab["accessory"].iloc[i] == 1:
            #Montando o gabarito:
            data["filename"].append("flipped_"+str(gab["filename"].iloc[i]))
            data["hat"].append(gab["hat"].iloc[i])
            data["veil"].append(gab["veil"].iloc[i])
            data["glasses"].append(gab["glasses"].iloc[i])
            data["accessory"].append(1)

            #flip da imagem:
            img = cv2.imread(f)
            flipped = np.flip(img[:, :, :], axis=1)
            #Salvando a imagem no dataset:
            fd = os.path.join("aumentation", "flipped_"+str(gab["filename"].iloc[i]))
            cv2.imwrite(fd, flipped)
    else:
        pass

df = pd.DataFrame(data)
display(df)
```
